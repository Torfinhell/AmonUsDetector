```python
import gdown
from pathlib import Path


GDRIVE_URLS = {
"models": {
"https://drive.google.com/uc?id=1e6uLmJ8UfKqala8XifAteTkd8aocxm1x": "checkpoints/fcos-epoch=02-loss_val=1.2543.ckpt",
}
}

def download_models(gdrive_urls):
    if "models" not in gdrive_urls:
        raise ValueError("No model URLs found")

    for url, dest_path in gdrive_urls["models"].items():
        dest_path = Path(dest_path)
        dest_path.parent.mkdir(parents=True, exist_ok=True)
        
        if dest_path.exists():
            print(f"File already exists: {dest_path}")
            continue
        
        print(f"Downloading {url} to {dest_path} ...")
        gdown.download(url, str(dest_path), quiet=False)
        print(f"Saved to {dest_path}")

download_models(GDRIVE_URLS)
```
