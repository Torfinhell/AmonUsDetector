````python
import xml.etree.ElementTree as ET

class SVGPath:
    def __init__(self, d, fill="#000000", fill_rule="nonzero", clip_rule=None):
        self.d = d
        self.fill = fill
        self.fill_rule = fill_rule
        self.clip_rule = clip_rule

    def to_svg(self):
        attrs = [
            f'd="{self.d}"',
            f'fill="{self.fill}"',
            f'fill-rule="{self.fill_rule}"'
        ]
        if self.clip_rule:
            attrs.append(f'clip-rule="{self.clip_rule}"')
        return f"<path {' '.join(attrs)} />"


class SVGImage:
    def __init__(self, width=None, height=None, viewbox=None):
        self.width = width
        self.height = height
        self.viewbox = viewbox
        self.shapes = []

    def add(self, shape):
        self.shapes.append(shape)

    def to_svg(self):
        body = "\n  ".join([shape.to_svg() for shape in self.shapes])

        svg_attrs = ['xmlns="http://www.w3.org/2000/svg"']

        if self.width:
            svg_attrs.append(f'width="{self.width}"')
        if self.height:
            svg_attrs.append(f'height="{self.height}"')
        if self.viewbox:
            svg_attrs.append(f'viewBox="{self.viewbox}"')

        return (
            f'<svg {" ".join(svg_attrs)}>\n'
            f'  {body}\n</svg>'
        )

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(self.to_svg())


# ------------------
# LOAD ORIGINAL SVG
# ------------------

tree = ET.parse("amon_us.svg")
root = tree.getroot()

# get original width/height/viewBox
orig_width = root.attrib.get("width")
orig_height = root.attrib.get("height")
orig_viewbox = root.attrib.get("viewBox")  # might be None

# find first <path>
namespace = {"svg": "http://www.w3.org/2000/svg"}
path_el = root.find(".//svg:path", namespace)
if path_el is None:
    raise ValueError("No <path> found in amon_us.svg")

path_data = path_el.attrib.get("d")
fill = path_el.attrib.get("fill", "#000")
fill_rule = path_el.attrib.get("fill-rule", "nonzero")
clip_rule = path_el.attrib.get("clip-rule")

p = SVGPath(
    d=path_data,
    fill=fill,
    fill_rule=fill_rule,
    clip_rule=clip_rule
)

# ------------------
# CREATE MATCHING SVG
# ------------------

svg = SVGImage(
    width=orig_width,
    height=orig_height,
    viewbox=orig_viewbox
)

svg.add(p)
svg.save("output.svg")
````

````python
import cv2
import numpy as np
import random
import cairosvg
from io import BytesIO
from PIL import Image
````

````python
def random_background(width=800, height=600):
    # Random color per pixel
    bg = np.random.randint(0, 256, (height, width, 3), dtype=np.uint8)
    return bg
````

````python
def svg_to_cv2_image(svg_instance, width, height):
    # Render SVG to PNG bytes
    png_bytes = cairosvg.svg2png(bytestring=svg_instance.to_svg().encode('utf-8'),
    output_width=width, output_height=height)
    # Load PNG bytes to PIL
    pil_image = Image.open(BytesIO(png_bytes)).convert("RGBA")
    # Convert PIL to OpenCV (BGR + alpha)
    cv_image = np.array(pil_image)
    # Split alpha and BGR channels
    bgr = cv_image[:, :, :3]
    alpha = cv_image[:, :, 3] / 255.0
    return bgr, alpha
````

````python
def overlay_image(bg, fg_bgr, fg_alpha, x, y):
    h, w = fg_bgr.shape[:2]
    bg_h, bg_w = bg.shape[:2]

    # Clip if the image goes out of bounds
    if x + w > bg_w:
        w = bg_w - x
        fg_bgr = fg_bgr[:, :w]
        fg_alpha = fg_alpha[:, :w]
    if y - h < 0:
        h = y
        fg_bgr = fg_bgr[-h:, :]
        fg_alpha = fg_alpha[-h:, :]

    # Overlay using alpha blending
    for c in range(3):
        bg[y-h:y, x:x+w, c] = (fg_alpha * fg_bgr[:, :, c] +
                                (1 - fg_alpha) * bg[y-h:y, x:x+w, c])
    return bg
````
